```python
# importing required modules
import PyPDF2
import re
import pandas as pd
from itertools import chain
```

```python
import io
 
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfpage import PDFPage
 
def extract_text_from_pdf(pdf_path):
    resource_manager = PDFResourceManager()
    fake_file_handle = io.StringIO()
    converter = TextConverter(resource_manager, fake_file_handle)
    page_interpreter = PDFPageInterpreter(resource_manager, converter)
 
    with open(pdf_path, 'rb') as fh:
        for page in PDFPage.get_pages(fh, 
                                      caching=True,
                                      check_extractable=True):
            page_interpreter.process_page(page)
 
        text = fake_file_handle.getvalue()
 
    # close open handles
    converter.close()
    fake_file_handle.close()
 
    if text:
        return text
 
```

```python
page_data_all = extract_text_from_pdf('ASA-2020-Catalog-web.pdf')
```

```python
regex_pattern = "ISBN.[0-9,\-]*"
match_data = re.findall(regex_pattern,page_data_all)
```

```python
match_data
```

```python
U_match_isbn = list(dict.fromkeys(match_data))
```

```python

# creating a pdf file object
pdfFileObj = open('ASA-2020-Catalog-web.pdf', 'rb')

# creating a pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

total_pages = pdfReader.numPages
```

```python
len(U_match_isbn)
len(match_data)
```

```python
regex_pattern = "ISBN [0-9,-]*.*\n.*.*"
all_match_data = []
for i in range(total_pages):
    pageObj = pdfReader.getPage(i)
    page_data = pageObj.extractText()
    match_data = re.findall(regex_pattern,page_data)
    all_match_data.append(match_data)
    
```

```python
all_match_data
```

```python
combined_list = list(chain(*all_match_data))
unique_isbn = list(dict.fromkeys(combined_list))
```

```python
combined_list
```

```python
data_v = []
for i in unique_isbn:
    isbn_v = re.findall('ISBN [0-9 -]*',i)
    part_v = re.findall('#.*',i)
    data_v.append([isbn_v,part_v])
    
```

```python
df = pd.DataFrame(data_v,columns=['ISBN','Part'])
```
